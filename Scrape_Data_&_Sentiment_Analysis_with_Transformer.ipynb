```python
import pandas as pd
import numpy as np
from google_play_scraper import Sort, reviews_all
import plotly.express as px
from transformers import pipeline
```

```python
hk_project = reviews_all(
    'id.go.kpu.sirekap2024',
    sleep_milliseconds=0,
    lang='id',
    country='ID',
    sort=Sort.NEWEST
)
```

```python
# hk_project
```

```python
df = pd.json_normalize(hk_project)
```

```python
df.head()
```

```python
df['score'].mean()
```

```python
df['reviewCreatedVersion'].value_counts()
```

```python
sentiment_analysis = pipeline("sentiment-analysis", model="ayameRushia/bert-base-indonesian-1.5G-sentiment-analysis-smsa")
```

```python
print(sentiment_analysis("Kamu jelek sekali"))
```

```python
df.dtypes
```

```python
df['content'] = df['content'].astype(str)
```

```python
df['result'] = df['content'].apply(lambda x: sentiment_analysis(x))
```

```python
df.head()
```

```python
df['sentiment']= df['result'].apply(lambda x: x[0]['label'])
df['score']= df['result'].apply(lambda x: x[0]['score'])
```

```python
df.head()
```

```python
flg = px.histogram(df, x='sentiment', color='sentiment', text_auto=True)
flg.show()
```

```python
df.to_csv('hk_project.csv')
```
